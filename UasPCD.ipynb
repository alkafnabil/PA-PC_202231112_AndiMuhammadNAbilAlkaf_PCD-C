```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'nanas.jpeg'
image = cv2.imread(image_path)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
lower_orange = np.array([10, 100, 100])
upper_orange = np.array([20, 255, 255])

lower_dark_green = np.array([35, 50, 20])
upper_dark_green = np.array([90, 255, 100])
```

```python
orange_mask = cv2.inRange(hsv, lower_orange, upper_orange)
dark_green_mask = cv2.inRange(hsv, lower_dark_green, upper_dark_green)
orange_segment = cv2.bitwise_and(image_rgb, image_rgb, mask=orange_mask)
dark_green_segment = cv2.bitwise_and(image_rgb, image_rgb, mask=dark_green_mask)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0, 0].imshow(image_rgb)
ax[0, 0].set_title('Original Image')
ax[0, 0].axis('off')

ax[0, 1].imshow(orange_mask, cmap='gray')
ax[0, 1].set_title('Orange Mask (Pineapple)')
ax[0, 1].axis('off')

ax[1, 0].imshow(orange_segment)
ax[1, 0].set_title('Segmented Pineapple (Orange)')
ax[1, 0].axis('off')

ax[1, 1].imshow(dark_green_segment)
ax[1, 1].set_title('Segmented Leaves (Dark Green)')
ax[1, 1].axis('off')

plt.tight_layout()
plt.show()
```
